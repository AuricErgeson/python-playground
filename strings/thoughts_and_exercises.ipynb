### Strings

strings are immutable sequence in python, you can do operations on the copy of the string but not the original.

each character in a string has an index which is good to access each character or in a group and do operatons


```python
message = "Hello my friend, will you come today?"


for index,character in enumerate(message):
  print(index,character)
```

#### 2 Count character frequency in a string.

```python
frequency = {}

for character in string:
  if character in frequency:
    frequency[character] += string.count(character)
  else:
    frequency[character]= string.count(character)

frequency
```

#### 3 Get string of first and last 2 chars.

```python
sample = "w3resource"
sample[:2]

def first_last(text):
  if len(text)> 3:
    first= text[:2]
    last = text[-2:]
    result = first+last
  elif len(text) == 2:
   result = text+text
  else:
    result = ""
  return result

first_last(sample)
first_last("Yes")
first_last("No")
```

#### 4 Replace first char occurrences with $.

```python
e4 = "eben"

first_cha= e4[0] #=e
repl = e4.replace(first_cha,"$") #= $b$n

first_cha + repl[1:]# ($:0 b :1 $:2 n:3)
```

#### 5 Add ing or ly to a string.

```python
def add_end(text):
  if len(text)>=3 and text.endswith("ing"):
    result = text+"ly"
  elif len(text) >= 3 and not text.endswith("ing"):
    result = text +"ing"
  else:
    result = text
  return result

add_end("no")
```

#### 9. Remove nth character from a string.

```python
st = "Nitonde"
list(st)

def remove_nt(text,n):
  txt_to_list=list(text)
  txt_to_list.pop(n)
  result = "".join(txt_to_list)

  return result

remove_nt(st,1)
```

#### 10 Remove odd index chars from a string

```python
def remove_od(text):
  text_to_list= list(text)
  for index, ch in enumerate(text_to_list):
    if index %2==0:
      text_to_list.pop(index)
  result = "".join(text_to_list)
  return result

remove_od("Nitonde")
```
